```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('unemployment analysis.csv.xls' , sep=",")
df1 = pd.read_csv('continents2.csv', sep=",")
df2 = pd.read_csv('owid-covid-data.csv', sep=",")
```

```python
df.set_index("Country Code")
```

```python
df1.set_index("alpha-3")
```

```python
dfresult = pd.merge(df, df1, how="left", left_on = "Country Code", right_on = "alpha-3")

dfresult.columns[:]
```

```python
dfresult[['Country Name', 'Country Code', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', 'region', 'sub-region',]]
```

```python
dfnew = pd.DataFrame()
for y in df.columns[ 2: ]:
    dftemp = dfresult.loc[ : , ["Country Name","Country Code", "region", "sub-region",y] ]
    dftemp["Year"] = y
    dftemp.columns = ['Country Name', 'Country Code', "region", "sub-region", 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'region', 'sub-region', 'Year', 'Value'] ]
    
    dfnew = pd.concat([dfnew, dftemp])
```

```python
listacode = list()
listacode = dfnew["Country Code"].unique().tolist()
```

```python
listageneral = list()

for i in listacode:
    resul = dfnew.where(dfnew["Country Code"] == i)
    resul.dropna(inplace = True)
    resultado = resul[["Country Name", "Country Code", "Year", "Value"]][resul.Value == resul["Value"].max()]
    listageneral.append(resultado)
```

```python
dfunemployment = pd.DataFrame()
for i in listageneral:
    dfunemployment = dfunemployment.append(i)
```

```python
dfunemployment.set_index("Country Code")
```

```python
df2 = df2[["iso_code", "date", "new_cases"]]

df2
```

```python
listaisocode = list()
listaisocode = df2["iso_code"].unique().tolist()
```

```python
listadf = list()

for i in listaisocode:
    resul = df2.where(df2["iso_code"] == i)
    resul.dropna(inplace = True)
    resultado = resul[["iso_code", "date", "new_cases"]][resul.new_cases == resul["new_cases"].max()]
    listadf.append(resultado)
```

```python
dfcovidcases = pd.DataFrame()
for i in listadf:
    dfcovidcases = dfcovidcases.append(i)
```
